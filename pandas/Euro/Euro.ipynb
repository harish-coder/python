### Step 14. Present only the Shooting Accuracy from England, Italy and Russia

```python
plot_accuracy_df=euro12[(euro12['Team'] == 'England') | (euro12['Team'] == 'Italy') | (euro12['Team'] == 'Russia')]
```

```python
%matplotlib notebook 
import matplotlib.pyplot as plt
plt.bar( plot_accuracy_df['Team'], plot_accuracy_df['Shooting Accuracy'], label="Example two", color='g')
plt.legend()
plt.xlabel('Team');
plt.ylabel('Shooting Accuracy');
plt.title('Team Vs Shooting Accuracy')
plt.show()
```
